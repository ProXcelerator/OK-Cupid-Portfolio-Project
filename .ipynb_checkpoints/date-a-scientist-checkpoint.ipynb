# Visualization of Outliers

# Age

```python
# visualize the age outliers
sns.displot(data=df, x='age', hue='sex', aspect=2, binwidth=5, multiple='stack')
```

This showes that more men use this app than women. The porpotion of age between the two genders are relitivly the same. Meaning more people in their late 20s to mid 30s are using the app compared to peole in their 40s to 70s. 

# Age with outliers removed

```python
# remove the outliers and replaced with nan values.
df['age'] = df[df.age<=69]['age']
print('There were', df['age'].isnull().sum(), 'outliers in the ages feature that we removed.')

# visualize the new feature range
sns.displot(data=df, x='age', hue='sex', aspect=2,  binwidth=5, multiple='stack')
```

This app is mostly used by middle aged people. Some older people use it but might not have as much success. 

# Height

```python
sns.displot(data=df, x='height', hue='sex', binwidth=3, multiple='stack')
```

The height feature shows it has outliers between 0-58 and between 81-100. Lets remove those and look at it again.

```python
df['height'] = df[df.height>=60][df.height<=80]['height']
sns.displot(data=df, x='height', hue='sex', binwidth=1, aspect=2, multiple='stack')
```

# Categorical Data

```python
for data in df.columns:
    if df[data].dtypes==np.dtype('O'):
        print(data, " - ", df[data].isnull().sum())
```

Most of the categorical features have missing values. 'last_online', 'location', 'orientation', 'sex', and 'status do not need to be adjusted for missing values. 

# Cleaning Features

```python
import re

# lets look at all the features that are catigorical but are not the essays. 
for i in df.columns:
    if not re.match(r"essay\B", i) and  df[i].dtype == np.dtype('O'):
        print('Column name:', i)
        print(df[i].value_counts(dropna=False))
        print(" ")
        print(" ")
```

```python
# we want the first word in each religion description to simplify the feature variables. 
df['religion'] = df['religion'].str.split().str.get(0)
print(df['religion'].value_counts(dropna=False))
```

```python
# replace &rsquo; for ' so we can understand the variables better.
df['offspring'] = df['offspring'].apply(lambda x : x.replace('&rsquo;', '\'') if type(x) == str else x)
print(df['offspring'].value_counts())
```

```python
# grab just the first word to simplify the feature variables
df['sign_cleaned'] = df['sign'].str.split().str.get(0)
print(df['sign_cleaned'].value_counts())
```

```python
# simplify this feature like we did previously 
df['religion_cleaned'] = df['religion'].str.split().str.get(0)
print(df['religion'].value_counts())
```

# Data Preprocessing
## After we proprocess the data we will look at if we can predict the sign of an individual when looking at different categorical and numarical features


```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import train_test_split
```

```python
# list of columns that we want to make a new dataframe from 
predictor_col = ['diet', 'orientation', 'pets', 'religion_cleaned','sex', 'job', 'smokes', 'drinks', 'drugs', 'income']
# create new dataframe
df2 = df[predictor_col]
```

```python
# change the features variables to binary individual features using One Hot Encoding
df_encoded = pd.get_dummies(df2, columns =predictor_col, prefix = predictor_col, dummy_na=True)
# add sign_cleaned to the dataframe
df_encoded['sign_cleaned'] = df['sign_cleaned']
# remove any nan values
df_encoded = df_encoded.dropna(subset=['sign_cleaned'])
# check if nan values have been removed
print(df_encoded.sign_cleaned.isna().sum())

```

```python
# create x and y variables for prediction models
x = df_encoded.drop('sign_cleaned', axis = 1)
y = df_encoded['sign_cleaned']
```

```python
# create train and test sets for prediction validation
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .25, random_state=7)
```

# Income Prediction- O-H-E

```python
pred_col = ['education', 'job', 'sex', 'location']
df3 = df[pred_col]
```

```python
df3 = pd.get_dummies(df3, columns=pred_col, prefix=pred_col, dummy_na=True)
df3['income'] = df['income']
df3['age'] = df['age']
```

```python
df3 = df3.dropna()
df3.income.isna().sum()
df3
```

```python
x = df3.drop('income', axis=1)
y = df3['income']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.25 , random_state=20)

lr = LogisticRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
print(lr.score(x_train,y_train))
print(accuracy_score(y_test, y_pred))
```
